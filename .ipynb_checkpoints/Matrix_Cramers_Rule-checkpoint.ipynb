```python
import copy
def Determinant_of_a_Matrix(matrix):    #[row][col]
    #varible initialization
    size = len(matrix) 
    sum_determination = 0
    temp_sum=0
    
    #Row 1 has all variable constant
    copy_matrix_first_row = copy.deepcopy(matrix)[0]   ### for list under list the copy will effect the original list so using deepcopy
    # print("\nRow 0      :\t",copy_matrix_first_row)

    #Rest of the matrix where we will find detereminant and also reduce it to small 
    copy_matrix1 = copy.deepcopy(matrix)   ### for list under list the copy will effect the original list so using deepcopy
    # print('matrix \n-------------------\n',copy_matrix1,'\n-------------------')
    # print('-------------------\n',copy_matrix1,'\n-------------------')
    
    len_row1 = len(copy_matrix_first_row)
    # print("no of col in the on working matrix:",len_row1)

    if len_row1 <2:              #only one row/col remain
        # print('last element',copy_matrix1[0][0])
        return copy_matrix1[0][0]
    else:                                               #[row][col]
        for y in range(0,len_row1,1):                                #col   --- variable 
            # print("\ny = " ,y ,'\t\t matrix y element is:', copy_matrix_first_row[y])
            copy_matrix1 = copy.deepcopy(matrix)                     ### for list under list the copy will effect the original list so using deepcopy
            del copy_matrix1[0]
            # temp_sum= copy_matrix_first_row[y]
            for z in range(0,len(copy_matrix1),1):                   #row control ---- eqn 
                for x in range(0,len(copy_matrix1[z]),1):            #coontrol element in eq
                    if y==x:
                        del copy_matrix1[z][x]
            
            # print('\nremaining matrix to decide if we need more recursion\n**************\n',copy_matrix1,'\n*******************')
            temp_sum = 0
            # print("temp_sum before/reset :" ,temp_sum)
            last_det = Determinant_of_a_Matrix(copy_matrix1)
            if y%2 ==0:
                temp_sum = copy_matrix_first_row[y] * last_det
            else:
                temp_sum = -1*copy_matrix_first_row[y] * last_det
            # print('y =',y,'\t\t',copy_matrix_first_row[y] ,' * ', last_det ,' = ',temp_sum , '  sum_determination = ',sum_determination)
            sum_determination = sum_determination + temp_sum
        # print("--------------------------------------------------------------------------------")
        # print('before\t sum_determination : ',sum_determination,'\t temp_sum :',temp_sum)
        
    # print('sum_determination : ',sum_determination,'\t temp_sum :',temp_sum)
    return sum_determination
#---------------------------------------------------------------------#
```

```python
'''
def trim_matrix(matrix , col):               # to find the 
    num_of_eqn = len(matrix)
    num_of_element = len(matrix[0])
    for i in range(0,num_of_eqn,1):
        for j in range(0,num_of_element,1):
            if j== col:
                del matrix[i][col]
    # print(matrix)
    return matrix
'''
#------------------------------------------------------------------------#
def matrix_separate(matrix):                         ## extract variable and const from the whole_matrix/equation
    # print(matrix)
    const_m = []
    num_of_eqn = len(matrix)
    num_of_element = len(matrix[0])
    # print("num_of_element",num_of_element , '\tnum_of_eqn',num_of_eqn)
    for i in range(0,num_of_eqn,1):
        for j in range(0,num_of_element,1):
            # print(i,'=i\tj=',j,'\t',matrix[i][j],end='')
            # print('\t',matrix)
            if j == num_of_eqn:
                const_m.append(matrix[i][j])
                del matrix[i][j]
    # print(matrix)
    return list(const_m),matrix
#------------------------------------------------------------------------#
def replace_variable(const ,variable_matrix,variable_position):
    # print('const:',const)
    # print('variabale',variable_matrix)
    num_of_eqn = len(variable_matrix)
    num_of_element = len(variable_matrix[0])
    # print("num_of_element",num_of_element , '\tnum_of_eqn',num_of_eqn)
    for i in range(0,num_of_eqn,1):
        for j in range(0,num_of_element,1):
            if j == variable_position:
                # print(variable_matrix[i][j],'\t',const[i])
                variable_matrix[i][j]=const[i]
            True
    # print(variable_matrix)
    return variable_matrix
#------------------------------------------------------------------------#    
```

```python
# two variable
import copy
#------------------------------------------------------------------------#
            #1 variable
whole_matrix =[[100,1]]
            #2 variable
# whole_matrix = [[12,-10,46],[3,20,-11]]
# whole_matrix = [[6.6,0.95,5.2],[4.2,8.6,19.3]]
            #3 variable
# whole_matrix = [[3,-4,8,34],[4,1,-2,1],[-6,-13,20,61]]
# whole_matrix = [[3,-8,10,8],[-1,10,9,-15],[2,-6,1,11]]
# whole_matrix = [[2,4,-6,19],[3,6,-9,30],[4,-7,1,15]]
whole_matrix = [[1,1,1,6],[5,-6,8,17],[2,3,-1,5]]
#------------------------------------------------------------------------#
print('eqn is in format ax1+bx2+cx3+.....=const ie [a,b,c.....,const]\n#------------------------------------------------------------------------#')
print("Test Matrix :\t",whole_matrix)
const ,variable_matrix = matrix_separate(whole_matrix)
D = Determinant_of_a_Matrix(trim_matrix(whole_matrix,-1))
print("Determinant_of_a_Matrix D:\t",D)

if D == 0:
    print("No solution")
else:
    print("#------------------------------------------------------------------------#\nsolutions are as follow\n#------------------------------------------------------------------------#")
    for i in range(0,len(variable_matrix),1):
        print(f'x{i} = ',end = '')
        c=copy.deepcopy(const)
        vm=copy.deepcopy(variable_matrix)
        ax = Determinant_of_a_Matrix(replace_variable(c ,vm,i))
        print(ax/D)
    
```
