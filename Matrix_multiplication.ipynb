```python
amatrix_a = [[7,14,15,6],[4,8,12,3],[14,21,6,9],[13,7,6,4]]
bmatrix_b = [[5,7,14,2],[8,16,4,9],[13,6,8,4],[6,3,2,4]]
sol =[[378,381,28,224],[258,237,190,140],[370,497,346,277],[223,251,266,129]]
#--------------------------------------------------------------#
def size_of_matrix(a):
    row = len(a)
    col = len(a[0])
    for i in range(0,row,1):
        # print(len(a[i]))
        if col < len(a[i]):
            col = len(a[i])
            return False
    # print(row ,'\t',col)
    return row , col
#--------------------------------------------------------------#
#getting size of matrix
amatrix_row , amatrix_col = size_of_matrix(amatrix_a)
bmatrix_row , bmatrix_col = size_of_matrix(bmatrix_b)
#--------------------------------------------------------------#
print("matrix A : ",amatrix_a)
print("matrix B : ",bmatrix_b)
print("desired sol",sol)
#--------------------------------------------------------------#
if amatrix_col != bmatrix_row:
    print("No solution ")
else:
    ab_multiple=[]
    for row in range (0,amatrix_row,1):
        row_matrix=[]
        for col in range(0,bmatrix_col,1):
            sum = 0
            for j in range(0,bmatrix_col,1):      
                aa = amatrix_a[row][j]
                bb = bmatrix_b[j][col]
                sum = sum + aa * bb
                # print(f'sum : {sum}  i = {i}  j = {j}  col = {col}  row = {row}   a = {aa}   b = {bb}  multi = {aa * bb}  sum = {sum}')
            row_matrix.append(sum)
            # print(sum)
        ab_multiple.append(row_matrix)
print("sol generate:",ab_multiple)
            
```
